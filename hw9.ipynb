## Programming Problems

```python
from scipy.stats import rv_discrete, expon, norm, probplot
from matplotlib import pyplot as plt
import numpy as np
```

## Problem 1

```python
sample_norm = norm(loc = 0, scale = 2)
seq = np.zeros(201)
seq[0] = 1

for x in range(1,201):
    seq[x] = 0.7 * seq[x-1] - 2.1 + sample_norm.rvs()

plt.plot(seq)
```

## Problem 2

```python
print(f"Mean: {seq[50:].mean()}\n\
Variance: {seq[50:].var()}")
```

The expectation above is almost identical to the theoretical one, but the variance is off. However, this can be due to the small sampling size.
